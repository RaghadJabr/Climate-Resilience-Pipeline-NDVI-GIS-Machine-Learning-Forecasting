# NLP Analysis (MVP)
This notebook demonstrates a minimal NLP pipeline on sample climate news to extract keywords per region and date.


```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
news = pd.read_csv('../data/raw/news_sample.csv')
news.head()
```

```python
tfidf = TfidfVectorizer(max_df=0.95, min_df=1, stop_words='english')
X = tfidf.fit_transform(news['text'])
nmf = NMF(n_components=2, random_state=0).fit(X)
terms = tfidf.get_feature_names_out()
for i, comp in enumerate(nmf.components_):
    top_idx = comp.argsort()[-8:][::-1]
    print(f'Topic {i}:', ', '.join(terms[top_idx]))
```
