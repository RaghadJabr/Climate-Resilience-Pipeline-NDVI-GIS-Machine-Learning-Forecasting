# ML Forecast (MVP)
Baseline regression to predict next-day NDVI anomaly from lagged climate features.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
df = pd.read_csv('../data/processed/features.csv', parse_dates=['date'])
# create next-day target within region
df = df.sort_values(['region','date'])
df['target_anom_next'] = df.groupby('region')['ndvi_anom'].shift(-1)
df = df.dropna()
features = [c for c in df.columns if any(k in c for k in ['temp_lag','precip_lag','ndvi_lag'])]
X = df[features]
y = df['target_anom_next']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
m = LinearRegression().fit(X_train, y_train)
pred = m.predict(X_test)
print('MAE:', mean_absolute_error(y_test, pred))
```
