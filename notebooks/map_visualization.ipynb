# Map Visualization (MVP)
Quick Folium map showing latest NDVI anomaly per region and NLP events overlay.

```python
import pandas as pd, folium
feat = pd.read_csv('../data/processed/features.csv', parse_dates=['date'])
news = pd.read_csv('../data/raw/news_sample.csv', parse_dates=['date'])
m = folium.Map(location=[31.95, 35.9], zoom_start=7)
latest = feat.sort_values('date').groupby('region').tail(1)
for _, row in latest.iterrows():
    folium.CircleMarker([row['lat'], row['lon']], radius=10,
                        tooltip=row['region'],
                        popup=f"NDVI anom: {row['ndvi_anom']:.3f}").add_to(m)
m
```
